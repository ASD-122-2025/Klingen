# 1 . Написати функцію pop_n(), що видаляє елементи стека з його початку до номера n включно.

```python
def pop_n(stack, n):
    for i in range(n):
        if stack:
            stack.pop()
        else:
            break
```

Приклад використання

```python
stack = [1, 2, 3, 4, 5]  
pop_n(stack, 3)
print(stack)  
```

# 3 . Написати функцію print_n(), що друкує елементи черги з його початку до номера n включно.

```python
def print_n(queue, n):
    for i in range(min(n, len(queue))):
        print(queue[i])
```

Приклад використання

```python
queue = ['яблуко', 'груша', 'банан', 'вишня', 'ківі']
print_n(queue, 3)
```

# 5 . Контрольні питання

* Що таке стек і які операції можна виконувати зі стеком? - Стек – це базова структура даних, в якій ми можемо тільки вставляти або видаляти елементи на початку стека.Існує три основні операції, які можуть виконуватися в стеках: вставка елемента в стек (push), видалення елемента з стека (pop) і відображення вмісту стека (pip).
* Яка основна відмінність між стеком та чергою? - Основна відмінність між стеком і чергою полягає в принципі, за яким елементи додаються та видаляються , у стеку перший додається та перший виходить , а у черзі додається в кінець та видаляється з верхівки.
* Як ви можете реалізувати стек за допомогою масиву і за допомогою зв’язаного списку? Які переваги та недоліки кожного підходу? - Переваги масиву - Простота реалізації.Швидкий доступ до елементів (за індексом).Ефективне використання пам'яті (якщо розмір стека відомий заздалегідь).Недоліки - Фіксований розмір, Неефективне використання пам'яті, якщо розмір стека сильно змінюється.Переваги черги - Динамічний розмір , Ефективні операції push та pop O(1).Недоліки черги - Більше витрат пам'яті на зберігання покажчиків.Доступ до елементів не такий швидкий, як у масиві (потрібно проходити по списку).Трохи складніша реалізація.
* Які є застосування стека та черги в програмуванні і реальному житті? - Стек у програмуванні - Виклик функцій , пошук у глибину (DFS) , Аналіз виразу.Стек у реальному житті - стопка тарілок або чогось іншого , історія пошуку у веб браузері.Черга у програмуванні - Планування завдань , обробка запитів , ширина пошуку (BFS).Черга у житті - черга у магазині , обслуговування клієнтів у ресторані.
