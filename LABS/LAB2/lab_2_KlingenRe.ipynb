```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Обмін елементів
                swapped = True
        if not swapped:  # Якщо не було жодного обміну, завершити сортування
            break
    return arr
```
