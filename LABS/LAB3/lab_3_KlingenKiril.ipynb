# 1 . Виконати теоретичні розрахунки асимптотичної складності алгоритмів сорутвання методами вставки та бульбашки.

1.Метод вставки : В найкращому випадку ми матимемо O(n) в цьому випадку ми вставимо число у самий початок не проходячи через інші елементи масиву.В найгіршому (найчастіше зустрічаємому випадку) буде O(n^2).Можемо прописати це так T(n) = (n-1)(c1+c2+c6)+((n(n-1))/2)*(c3+c4+c5).Домінючим членом є n^2.
2.Метод бульбашки : Взагалом метод схожий на метод вставки але прописується він інакше T(n) = ((n-1)*c1)+ ((n(n-1))/2)*c2+c3+c4).O(n^2) - найгірший та в одночас найпоширеніший випадок.

# 2 . Реалізувати на Python алгоритм сортування методом бульбашки.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

# 3 . Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей , згенерувавши відповідні цілочисельні масиви.

```python
import time
import random
import matplotlib.pyplot as plt

# Сортування вставками
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >=0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Сортування бульбашкою
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Функція для вимірювання часу
def measure_time(sort_func, arr):
    start = time.perf_counter()
    sort_func(arr.copy())  # копіюємо масив, щоб не змінювати оригінал
    end = time.perf_counter()
    return end - start

# Масиви різної довжини
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]
times_insertion = []
times_bubble = []

for size in sizes:
    arr = [random.randint(0, 10000) for _ in range(size)]
    times_insertion.append(measure_time(insertion_sort, arr))
    times_bubble.append(measure_time(bubble_sort, arr))

# Побудова графіків
plt.plot(sizes, times_insertion, label='Insertion Sort', marker='o')
plt.plot(sizes, times_bubble, label='Bubble Sort', marker='s')
plt.xlabel('Розмір масиву')
plt.ylabel('Час виконання (секунди)')
plt.title('Порівняння часу сортування: вставки vs бульбашка')
plt.legend()
plt.grid(True)
plt.show()
```
