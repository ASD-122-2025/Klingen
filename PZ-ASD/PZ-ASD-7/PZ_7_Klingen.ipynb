# Мета : набути практичних навичок застосування базових алгоритмів на рядках та оцінювання їх асимптотичної складності.

# 1 . Маємо дві короткі послідовності символів: «HELLO» і «WORLD». Знайти найдовшу спільну підпослідовності символів, використовуючи алгоритм Хаббарда.


# 2 . Контрольні питання.

*  У чому полягає задача знаходження найдовшої спільної підпослідовності (LCS)? - Вона полягає в пошуку найбільшої за довжиною послідовності, яка є 
підпослідовністю двох або більше заданих рядків.
   
* Які головні методи можна використовувати для знаходження найдовшої спільної підпослідовності? - Динамічне програмування , Рекурсивний алгоритм , алгоритм Хаббарда , алгоритм повного перебору.
  
* Як працює алгоритм динамічного програмування для знаходження LCS? - він працює так : створюється таблиця для зберігання проміжних результатів , порівнюється кожен символ з символом іншої послідовності. Кожна клітинка таблиці залежить від попередніх, використовуючи збіг останніх елементів або максимум з сусідніх значень.

* Як працює алгоритм Хаббарда для знаходження LCS? - він працює схожим чином , але ми повинні розділити одну з послідовностей на дві частини та обчислити LCS порізно при чому для 2 частини ми інверсуємо 2 і 2 частину 1 послідовності  , після чого шукаємо точку поділу та знаходимо результат.

* Які переваги та недоліки алгоритмів динамічного програмування та Хаббарда для знаходження LCS? - переваги LCS - простота реалізації та зрозумілість , недоліки LCS - висока складність та надмірні обчислення.Переваги Хаббарда - масштабованість та використання рекурсивного поділу , недоліки - складніша реалізація , велика кількість обчислень.

* Які існують практичні застосування для задачі знаходження найдовшої спільної підпослідовності? - Виявлення плагіату , авто корекція тексту , виокремлення ключових фрагментів у тексті.
